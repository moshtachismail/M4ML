```python
import matplotlib.pyplot as plt

# Data
students = ['A', 'B', 'C']
X = [2.1, 2.9, 4.2]  # Hours studied
Y = [1.33, 2.66, 3.99]  # Predicted grades
true_Y = [2.1, 2.9, 4.2]  # True grades

# Calculate errors
errors = [true - pred for true, pred in zip(true_Y, Y)]

# Plot predictions vs true values
plt.figure(figsize=(8, 5))
plt.scatter(X, true_Y, color='blue', label='True Grade')
plt.scatter(X, Y, color='red', label='Predicted Grade')
plt.plot(X, Y, color='red', linestyle='--', alpha=0.5)

# Annotate with error values
for i in range(len(students)):
    plt.text(X[i], Y[i] + 0.1, f'Err={errors[i]:.2f}', ha='center', fontsize=9)

# Labels and legend
plt.title("Predicted vs True Grades")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
